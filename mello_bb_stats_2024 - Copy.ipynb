# Next, Let's Do Some Cleaning!

I realized something after that first game: I not only want to take statistics on how many points each player scored, but I also want to see how many points they attempted to score as well!

From that point forward, I began recording the attempted and made baskets for both 2 and 3 points, as well as the offensive and defensive rebounds. Because I did not start this until our Week2 game, from this point forward, we will only be looking at data starting at Week2. Since the Week1 data is incomplete, it might lead to incomplete results, and in data science, it is always important to be working with **good quality** data.

**First**, let's condense some of this data by creating some new columns!

```python
# Create 4 columns for each type of total by adding each week together!

df["Total 2-Point Made"] = df["Week2 2-Point Make"] + df["Week4 2-Point Make"] + df["Week5 2-Point Make"] + df["Week6 2-Point Make"]  + df["Week8 2-Point Make"] + df["Week9 2-Point Make"] + df["Play-Off 2-Point Make"]
df["Total 2-Point Attempted"] = df["Week2 2-Point Attempt"] + df["Week4 2-Point Attempt"] + df["Week5 2-Point Attempt"] + df["Week6 2-Point Attempt"]  + df["Week8 2-Point Attempt"] + df["Week9 2-Point Attempt"] + df["Week2 2-Point Make"] + df["Week4 2-Point Make"] + df["Week5 2-Point Make"] + df["Week6 2-Point Make"]  + df["Week8 2-Point Make"] + df["Week9 2-Point Make"] + df["Play-Off 2-Point Attempt"] + df["Play-Off 2-Point Make"]
df["Total 3-Point Made"] = df["Week2 3-Point Make"] + df["Week4 3-Point Make"] + df["Week5 3-Point Make"] + df["Week6 3-Point Make"]  + df["Week8 3-Point Make"] + df["Week9 3-Point Make"] + df["Play-Off 3-Point Make"]
df["Total 3-Point Attempted"] = df["Week2 3-Point Attempt"] + df["Week4 3-Point Attempt"] + df["Week5 3-Point Attempt"] + df["Week6 3-Point Attempt"]  + df["Week8 3-Point Attempt"] + df["Week9 3-Point Attempt"] + df["Week2 3-Point Make"] + df["Week4 3-Point Make"] + df["Week5 3-Point Make"] + df["Week6 3-Point Make"]  + df["Week8 3-Point Make"] + df["Week9 3-Point Make"] + df["Play-Off 3-Point Attempt"] + df["Play-Off 3-Point Make"]
```

```python
# List and the DataFrame of the total for made and attempt for 2 and 3 point shots

total_list = ["Mellophones", "Total 2-Point Made", "Total 2-Point Attempted", "Total 3-Point Made", "Total 3-Point Attempted"]
df_total_list = df[total_list]
df_total_list
```

That looks so much nicer, now we have the total number of attempts and makes for both 2 and 3-point shots!

**Now** let's find the **count totals** using multiplication!

```python
# Create 2 columns of the total point counts made for 2 and 3 point shots

df["2-Point Made Count"] = df["Total 2-Point Made"] * 2
df["3-Point Made Count"] = df["Total 3-Point Made"] * 3
```

```python
# List and the DataFrame of the total point count for 2 and 3 point shots

count_list = ["Mellophones", "2-Point Made Count", "3-Point Made Count"]
df_count_list = df[count_list]
df_count_list
```

**Finally**, lets use the `df_total_list` to find the **percentage** of the likelihood of each player making a 2 or 3 point shot!

```python
# Create 2 columns for 2 and 3 point shot percentages

df["2-Point Made Percent"] = (df["Total 2-Point Made"] / df["Total 2-Point Attempted"]) * 100
df["3-Point Made Percent"] = (df["Total 3-Point Made"] / df["Total 3-Point Attempted"]) * 100
```

```python
# List and the Dataframe of the point percentages for 2 and 3 point shots

percent_list = ["Mellophones", "2-Point Made Percent", "3-Point Made Percent"]
df_percent_list = df[percent_list]
df_percent_list
```

# Player Analysis
You may have noticed from looking at the parts where we use `.nlargest` that there is one name in particular that stands out, and that is **Danny Young!** "Danny is one of our best players, but let's see how accurate he really is. Let's create a function that tells us how many 2-point and 3-point baskets Danny will make given a certain number of attempts!"

**Recall:**
- Danny's 2-Point Percent: **36.585366%**
- Danny's 3-Point Percent: **20.000000%**

```python
# Let's initialize some variables

danny_2_point_percent = 36.585366
danny_3_point_percent = 20.000000
danny_attempts = 40
```

```python
# Create a function that calculates shots a player makes given a certain number of attempts and their make percentage

def calculate_baskets_made(attempts, shooting_percentage):
    baskets_made = attempts * (shooting_percentage / 100)
    return baskets_made
```

```python
# Let's run the function for Danny's 2-point shot percentage

print("If Danny shoots " + str(danny_attempts) + " times, he can make " + str(calculate_baskets_made(danny_attempts, danny_2_point_percent)) + " 2-point baskets")
```

```python
# Let's run the function for Danny's 3-point shot percentage

print("If Danny shoots " + str(danny_attempts) + " times, he can make " + str(calculate_baskets_made(danny_attempts, danny_3_point_percent)) + " 3-point baskets!")
```

Now that we've found our best player's percent, lets see if we can find our **lowest player's statistics** (that is not 0)

```python
# Dataframe with the mellophones that have a 2-point percentage that is not zero

df_2_percent_list_not_zero = df_percent_list[df_percent_list["2-Point Made Percent"] > 0]
df_2_percent_list_not_zero
```
